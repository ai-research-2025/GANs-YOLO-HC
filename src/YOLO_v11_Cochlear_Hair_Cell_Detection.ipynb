**Comparision between Actual and Predicted objects**

```python
from os import name
import cv2
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Define colors for each class (example with 4 classes)
colors = [
    (255, 255, 255),   # Class 0: White  IHC
    (255, 0, 0),   # Class 1: OHC-1
    (0, 255, 0),   # Class 2:   OHC-2
    (0, 0, 255) # Class 3: OHC-3
]

# Define class names (example with 5 classes)
class_names = ['IHC', 'OHC-1', 'OHC-2', 'OHC-3']

color_codes = {name: color for name, color in zip(class_names, colors)}

# Select split Train/Test
split = 'Test'
# Paste image file name
image_file = '23_sample14_folder1_kaylee_20x_split_0_1.png'
image_data_path = '/content/output_320/images/' + split + '/' + image_file
annotation_data_path = '/content/output_320/labels/' + split + '/' + image_file[:-4] + '.txt'

# Open the image file
with open(image_data_path, 'rb') as f:
    image_data = f.read()

# Open the annotation file
with open(annotation_data_path, 'r') as f:
    annotation_data = f.read()

# Convert the image data to an image array
image = cv2.imdecode(np.frombuffer(image_data, np.uint8), cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_height, image_width = image.shape[:2]

# Parse the annotation data
annotations = []
for line in annotation_data.strip().split('\n'):
    parts = line.split()
    cls_id, x_center, y_center, width, height = map(float, parts)
    annotations.append((int(cls_id), x_center, y_center, width, height))

# Create copies of the image for ground truth and predictions
image_gt = image.copy()
image_pred = image.copy()

# Draw ground truth bounding boxes
for cls_id, x_center, y_center, width, height in annotations:
    x = int((x_center - width / 2) * image_width)
    y = int((y_center - height / 2) * image_height)
    w = int(width * image_width)
    h = int(height * image_height)
    color = colors[cls_id % len(colors)]
    cv2.rectangle(image_gt, (x, y), (x + w, y + h), color, 2) # 2px thickness

# Draw predictions bounding boxes
results = model(image)

# Extract bounding boxes, classes, names, and confidences
boxes = results[0].boxes.xyxy.tolist()
classes = results[0].boxes.cls.tolist()
names = results[0].names
confidences = results[0].boxes.conf.tolist()

# Draw bounding boxes and labels of detected objects
for i in range(len(boxes)):
    box = boxes[i]
    class_name = names[classes[i]]
    color = color_codes[class_name]
    # Draw bounding box
    cv2.rectangle(image_pred, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), color, 2)


# Create legend patches with RGBA colors
legend_patches = [mpatches.Patch(color=np.array(color)/255, label=class_names[i]) for i, color in enumerate(colors)]

# Display the images with bounding boxes in separate windows
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Ground truth
ax1.imshow(image_gt)
ax1.set_title('Ground Truth')
ax1.axis('off') # Hide the axis

# Predictions
ax2.imshow(image_pred)
ax2.set_title('Model Predictions')
ax2.axis('off') # Hide the axis
# show color legend
fig.legend(handles=legend_patches, loc='upper center', ncol=len(class_names))

plt.show()
```
